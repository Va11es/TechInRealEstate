```python
import pandas as pd

mean_days_df = pd.read_csv('cleaned sheets/Metro_mean_days_to_close.csv')
median_days_df = pd.read_csv('cleaned sheets/Metro_median_days_to_close.csv')
```

```python
print("Missing values in mean_days_df:")
print(mean_days_df.isnull().sum())

print("Missing values in median_days_df:")
print(median_days_df.isnull().sum())

mean_days_df.dropna(inplace=True)
median_days_df.dropna(inplace=True)
```

```python
print("Summary Statistics for Mean Days Dataset:")
print(mean_days_df.describe())

print("Summary Statistics for Median Days Dataset:")
print(median_days_df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of region names to compare
regions_to_compare_mean = [
    'United States',
    'New York, NY',
    'Los Angeles, CA',
    'Chicago, IL',
    'Dallas, TX',
    'Houston, TX',
    # Add more regions as needed
]

# Filter DataFrame for mean days
mean_days_df_filtered = mean_days_df[mean_days_df['RegionName'].isin(regions_to_compare_mean)]

# Assuming that the date columns start from the 6th column (index 5) to the end
date_columns_mean = mean_days_df.columns[5:]

# Plotting line chart for mean days to close
plt.figure(figsize=(12, 6))
for index, row in mean_days_df_filtered.iterrows():
    plt.plot(date_columns_mean, row[5:], label=row['RegionName'])

plt.title('Mean Days to Close Over Time')
plt.xlabel('Time')
plt.ylabel('Mean Days')
plt.legend()
plt.xticks(range(0, len(date_columns_mean), 6), date_columns_mean[::6], rotation=45)
plt.tight_layout()
plt.show()

# List of region names to compare
regions_to_compare_median = [
    'United States',
    'New York, NY',
    'Los Angeles, CA',
    'Chicago, IL',
    'Dallas, TX',
    'Houston, TX',
    # Add more regions as needed
]

# Filter DataFrame for median days
median_days_df_filtered = median_days_df[median_days_df['RegionName'].isin(regions_to_compare_median)]

# Assuming that the date columns start from the 6th column (index 5) to the end
date_columns_median = median_days_df.columns[5:]

# Plotting line chart for median days to close
plt.figure(figsize=(12, 6))
for index, row in median_days_df_filtered.iterrows():
    plt.plot(date_columns_median, row[5:], label=row['RegionName'])

plt.title('Median Days to Close Over Time')
plt.xlabel('Time')
plt.ylabel('Median Days')
plt.legend()
plt.xticks(range(0, len(date_columns_median), 6), date_columns_median[::6], rotation=45)
plt.tight_layout()
plt.show()
```
